Selecting specific values on DataFrame is very important in data operation to achieve your goal.

```python
# 2 ways to access the column
# using indexing or accessing it as attribute
df.country #attribute
df['country'] #indexing

df['country'][0]
df.iloc[0]
df.iloc[0:, 1]

df.loc[0, ['country', 'title', 'winery', 'price']]

# df.set_index('title')
```

Apparently in indexing pandas there are two ways to access data values
1. .iloc[]
2. .loc[]

.iloc is an index-based selection – [0:4] means 0,1,2,3
.loc is a label-based selection – it means it doesn't follow the counting convention of rang. [0:5] is counted as 0,1,2,3,4,5 that means you'll get 6 values

Another thing there's another feature of selections where the first coma is a row slector while the second coma iss the column selector
.iloc[0, 1] - first row, second column
.loc[0, ['country']] - can type string
